```python
import pandas as pd
import os

# List all files to check the uploaded file
print(os.listdir('/content/'))

# After finding the correct file name
file_path = '/content/HR Data.csv'  # Replace with your file's actual name
df = pd.read_csv(file_path)

# Preview the dataset
df.head()
```

```python
# Check the first few rows of the dataset
df.head()

# Check for information on data types and non-null values
df.info()

# Check for statistical summary of the numerical columns
df.describe()
```

```python
# Rename columns (replace with appropriate column names)
df = df.rename(columns={
    'Attrition': 'Yes/No',
    'DistanceFromHome': 'Distance to Home',
    'NumCompaniesWorked' : ' no of Past Companies',
    'PercentSalaryHike': '%salaryhike',
    'StockOptionLevel': 'SOL',
    'TotalWorkingYears':'no of workig yrs'
})
```

```python
# Check for missing values in each column
df.isnull().sum()
```

```python
df = df.dropna()  # Remove rows with missing data
```

```python
# Check data types of each column
df.dtypes
```

```python
# Remove duplicate rows
df = df.drop_duplicates()
```

```python
# Sort the column in ascending order
df['Age'] = df['Age'].sort_values(ascending=True)

# Sort the column in descending order
df['Age'] = df['Age'].sort_values(ascending=False)
```

```python
# Sort the column in ascending order
df['DailyRate'] = df['DailyRate'].sort_values(ascending=True)

# Sort the column in descending order
df['DailyRate'] = df['DailyRate'].sort_values(ascending=False)
```

```python
df.to_csv('/content/cleaned_HR_data.csv', index=False)
```
